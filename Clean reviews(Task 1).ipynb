```python
import numpy as np
import pandas as pd
```

```python
Data= pd.read_csv("BA_reviews.csv")
Data
```

```python
Data.drop("Unnamed: 0",axis=1,inplace=True)
Data
```

```python
# iterate over each element in the 'reviews' column and clean the data
for index, row in Data.iterrows():
    clean_review = row['reviews'].replace('✅ Trip Verified | ', '')
    Data.at[index, 'clean_reviews'] = clean_review

# print the updated DataFrame
print(Data)
```
